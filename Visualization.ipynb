```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import pickle
import numpy as np
sns.set(rc={'figure.figsize':(15,10)})
```

```python
df = pd.DataFrame(columns=['index_g','id_epoch','losses_validation'])
final_df = pd.DataFrame(columns=['index_g','id_epoch','losses_validation'])
for i in range(1,9):
    
    index = list(range(1,101))
    name = 'losses_validation'+str(i)+'.pkl'
    file = open('models_grid/'+name,'rb')
    loss = pickle.load(file)
    file.close()
    ids = [i]*len(loss)
    ids = [str(j) for j in ids]
    
    df['id_epoch'] = ids
    df['losses_validation'] = loss
    df['index_g'] = index
    
    final_df = final_df.append(df)
```

```python
final_df = final_df.reset_index(drop=True)
ax = sns.lineplot(x="index_g", y="losses_validation",hue="id_epoch", data=final_df,palette=sns.color_palette("Set1", 8))
```
